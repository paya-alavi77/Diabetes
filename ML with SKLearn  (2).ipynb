</p> 


````python
iris = datasets.load_iris()
````

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
Scikit-learn این داده‌ها را به صورت یک دیکشنری ذخیره می‌کند. اصلی‌ترین بخش‌ها (کلیدها)ی این داده عبارتند از:
</p> 

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
<ul style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
    <li> featture_names: حاوی نام ویژگی‌های داده است (نام ستون‌ها)</li>   
    <li> data: بخش اصلی داده که برای آموزش یک مدل استفاده می‌شود.</li> 
    <li> target_names: حاوی نام برچسب‌های داده است (نام‌رده‌ها)</li>   
    <li> target: مقدار برچسب هر نمونه داده</li>
</ul>
</p> 



````python
print("Features: ", iris.feature_names, "\n")
print("Targets(Labels): ", iris.target_names)

# print(iris.data)
# print(iris.target)
````

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
برای استفاده از این داده‌ها و بررسی راحت‌تر می‌توانیم یک DataFrame از آن درست کنیم:
</p> 


````python
import pandas as pd  

# Creating the dataframe of the dataset  
irisDF = pd.DataFrame(iris.data, columns = iris.feature_names) 
irisDF['Target'] = iris.target
irisDF['Label'] = irisDF['Target'].apply(lambda x: iris.target_names[x])


print(irisDF.shape, '\n')
irisDF.head()
````

````python
irisDF.tail()
````

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
برای درک بهتر داده‌ها می‌توان نصب طول گلبرگ به طول کاسبرگ را برای هر نمونه داده به نمایش گذاشت:
</p> 


````python
from matplotlib import pyplot as plt


plt.figure(figsize=(5, 4))
scatter= plt.scatter(irisDF["sepal length (cm)"], irisDF["petal length (cm)"],
                    c = irisDF["Target"])


plt.legend(handles = scatter.legend_elements()[0], 
           labels = ['setosa','versicolor','virginica'],
           loc="lower right", 
           title="Iris Type")


plt.xlabel("sepal length (cm)")
plt.ylabel("petal length (cm)")


plt.tight_layout()
plt.grid()
````

<h1  style="color:blue; font-family:Nazanin;" dir="rtl">ایجاد مجموعه داده‌های آموزشی، اعتبارسنجی و آزمون </h1>

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
برای شروع فرآیند یادگیری ابتدا باید داده‌ها را به مجموعه‌های آموزشی، اعتبارسنجی و آزمون تقسیم‌بندی کنیم.
</p> 

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
برای این‌کار از تابع زیر استفاده می‌کنیم:
</p> 

**```X_train, X_test, y_train, y_test = train_test_split(X, y,  test_size, random_state)```**

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
داده را به‌صورت تصادفی مجموعه‌های آموزشی و آزمون تقسیم‌بندی می‌کند.</p> 

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
<ul style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
    <li> X:  DataFrame حاوی مقادیر اصلی داده</li>   
    <li> y: بردار حاوی مقادیر برچسب‌ها</li> 
    <li> test_size: درصد داده‌ای که باید در مجموعه آزمون قرار گیرد</li>   
    <li> random_state: دانه (seed) ابتدایی برای تولید عدد تصادفی</li>
</ul>
</p> 


````python
# Importing the class to perform train test split from model_selection module  
from sklearn.model_selection import train_test_split  
````

````python
  
# Separating the dependent and independent features  
X = irisDF.iloc[:, :-2].values  # از ستون اول تا دوتا مانده به انتها
y = irisDF.iloc[:, -1].values   # آخرین ستون
  
# Creating a testing dataset of size 0.2 times the whole dataset  
X_train, X_test, y_train, y_test = train_test_split(  X, y, 
                                                    test_size = 0.2, 
                                                    random_state = 10
                                                   )  

# Creating a testing dataset of size 0.2 times the whole dataset  
X_train, X_val, y_train, y_val = train_test_split(  X_train, y_train, 
                                                    test_size = 0.2, 
                                                    random_state = 10
                                                   )  

# Printing the shape of the training and testing dataset  
print("Shape of training set:")
print(X_train.shape)  
print(y_train.shape)  

print("Shape of validation set:")
print(X_val.shape)  
print(y_val.shape)  

print("Shape of test set:")
print(X_test.shape)  
print(y_test.shape)  
````

<a id = "knn"></a>
<h1  style="color:blue; font-family:Nazanin;" dir="rtl">K-نزدیک‌ترین همسایه (KNN) </h1>

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
برای رده‌بندی داده‌های با استفاده از روش KNN ابتدا به سادگی و با استفاده از تابع زیر یک مدل اولیه KNN ایجاد می‌کنیم:
</p> 

**```model = KNeighborsClassifier(n_neighbors)```**


<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
<ul style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
    <li> n_neighbors:  مقدار K یا همان تعداد نزدیک‌ترین همسایه</li>   
</ul>
</p> 


````python
from sklearn.neighbors import KNeighborsClassifier  

knn = KNeighborsClassifier(n_neighbors = 3) 
````

<h2 style="color:red; font-family:Nazanin;" dir="rtl">آموزش</h2>

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
حال با ارسال مقادیر X, y که همان داده آموزشی هستند به تابع fit به‌راحتی می‌توان یک مدل را آموزش داد:</p> 

**```model.fit(X, y)```**


<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
<ul style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
    <li> X:  مقادیر ویژگی‌ها</li>  
    <li> y:  بردار حاوی برچسب‌ها</li>  
</ul>
</p> 


````python
knn.fit(X_train, y_train)
````

<h2 style="color:red; font-family:Nazanin;" dir="rtl">آزمون</h2>

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
با استفاده از تابع predics می‌توان خروجی مدل را برای مقادیر جدید به‌دست آورد:
</p> 

**```y_pred = model.predict(X)```**


<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
<ul style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
    <li> X:  DataFrame حاوی مقادیر جدید (داده‌های آزمون)</li>   
</ul>
</p> 


````python
y_pred_val = knn.predict(X_val)

y_pred = knn.predict(X_test)
````

````python
y_pred
````

````python
y_test
````

<h2 style="color:red; font-family:Nazanin;" dir="rtl">ارزیابی مدل</h2>

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
و درنهایت با استفاده از یکی از معیارهای ارزیابی که در کتابخانه sklearn.metrics فراهم آمده‌اند می‌توان عملکرد مدل را بررسی نمود. 
به عنوان مثال تابع زیر، مقدار دقت پیش‌بینی مدل بر روی داد‌های جدید را بررسی می‌کند:</p> 

**```accuracy_score(y_true, y_pred)```**


<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
<ul style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
    <li> y_true: بردار حاوی برچسب‌های حقیقی و درست</li>  
     <li> y_pred: بردار حاوی برچسب‌های پیش‌بینی شده</li>
</ul>
</p> 


````python
from sklearn.metrics import accuracy_score  

# Finding accuracy by comparing actual response values(y_test) with predicted response value(y_pred)  
print("Validation set accuracy:", accuracy_score(y_val, y_pred_val))  

print("Test set accuracy:", accuracy_score(y_test, y_pred)) 
````

````python
# Providing sample data and the model will make predictions out of that data  
  
sample = [[5, 5, 3, 4], [1, 10, 3, 5]]

preds = knn.predict(sample)  
print(preds)
````

<h2 style="color:red; font-family:Nazanin;" dir="rtl">انتخاب بهترین پارامتر (K)</h2>

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
برای انتخاب پارامتر بهینه K می‌توانیم با استفاده از اعتبارسنجی متقابل (cross validation) باتغییر مقدار K و بررسی عملکرد مدل بر روی مجموعه اعتبارسنجی به ازای هر K بهینه‌ترین مقدار K را به دست آوریم.
</p> 

````python
acc = []
for i in range(1,11):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_val)
    acc.append(accuracy_score(y_val, pred_i))

acc
````

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
در اینجا در یک حلقه، مقادیر K از ۱ تا ۱۰ تغییر کرده و به ازای هر K دقت مدل در رده‌بندی داده‌های اعتبارسنجی محاسبه می‌شود.</p>

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
    برای مشاهده‌ی بهتر، می‌توان این مقادیر را به ازای هر K رسم نمود.
</p> 

````python
plt.figure(figsize=(5, 4))

plt.plot(range(1, 11), acc, marker = "o")

plt.xticks(range(1,11))
plt.xlabel("K")
plt.ylabel("Validation set Accuracy")

plt.tight_layout()
plt.grid()
````

<h1  style="color:blue; font-family:Nazanin;" dir="rtl">رگرسیون لاجیستیک (Logistic Regression) </h1>

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
برای ایجاد یک مدل رگرسیون لاجیستیک به‌منظور رده‌بندی، از کلاس  <a href = "https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#sklearn.linear_model.LogisticRegression">LogisticRegression</a> از زیرکتاب‌خانه <a href="https://scikit-learn.org/stable/modules/classes.html#module-sklearn.linear_model"> sklearn.linear_model </a> استفاده می‌کنیم.
</p>  

````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score  

logit = LogisticRegression(random_state=0)

logit.fit(X_train,y_train)

y_pred = logit.predict(X_test)

print("\n Test set accuracy:", accuracy_score(y_test, y_pred)) 
````

<h1  style="color:blue; font-family:Nazanin;" dir="rtl">شبکه عصبی: پرسپترون چندلایه (Multi Layer Perceptron) </h1>

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
کتاب‌خانه SKlearn، تنها شبکه‌ی عصبی <a href = "https://scikit-learn.org/stable/modules/neural_networks_supervised.html">پرسپترون چند‌لایه</a> را پیاده‌سازی کرده و دراختیار کاربران قرارداده است.
</p>

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
    برای ایجاد یک مدل رده‌بندی می‌توان از کلاس <a href ="https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html">MLPClassifier</a> استفاده نمود. مهم‌ترین پارامترهای تابع سازنده این کلاس در ادامه معرفی می‌شوند. همچنین برای استفاده از شبکه‌های عصبی در یک مسئله رگرسیون می‌توان از کلاس <a href ="https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPRegressor.html#sklearn.neural_network.MLPRegressor">MLPRegressor</a> استفاده نمود.
    هردوی این کلاس‌ها بخشی از زیرکتابخانه‌ی <a href="https://scikit-learn.org/stable/modules/classes.html#module-sklearn.neural_network"> sklearn.neural_network </a> هستند. 
</p>



**```model = MLPClassifier(hidden_layer_sizes = (4, 5),
                           activation = function,
                           solver = 'adam')```**


<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
<ul style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
    <li> hidden_layer_sizes: تعداد لایه‌های میانی و واحدها (نورون‌ها)ی هرلایه را مشخص می‌کند.   
.</li>   
    <li> activation: تابع فعال‌سازی به‌کار رفته در شبکه‌عصبی را مشخص می‌کند.</li>   
    <li> solver: الگوریتمی که برای پیدانمودن نقطه بهینه تابع  هزینه به‌کار می‌رود را مشخص می‌کند.</li>   
    
</ul>
</p> 


````python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score  

nn = MLPClassifier(hidden_layer_sizes = (10,15),
                   activation = 'relu',
                   solver = 'adam')

nn.fit(X_train, y_train)

y_pred = nn.predict(X_test)


print("\n Test set accuracy:", accuracy_score(y_test, y_pred)) 
````
