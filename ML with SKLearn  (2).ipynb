````python
import pandas as pd  

# Creating the dataframe of the dataset  
irisDF = pd.DataFrame(iris.data, columns = iris.feature_names) 
irisDF['Target'] = iris.target
irisDF['Label'] = irisDF['Target'].apply(lambda x: iris.target_names[x])


print(irisDF.shape, '\n')
irisDF.head()
````

````python
irisDF.tail()
````

````python
from matplotlib import pyplot as plt


plt.figure(figsize=(5, 4))
scatter= plt.scatter(irisDF["sepal length (cm)"], irisDF["petal length (cm)"],
                    c = irisDF["Target"])


plt.legend(handles = scatter.legend_elements()[0], 
           labels = ['setosa','versicolor','virginica'],
           loc="lower right", 
           title="Iris Type")


plt.xlabel("sepal length (cm)")
plt.ylabel("petal length (cm)")


plt.tight_layout()
plt.grid()
````

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
<ul style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
    <li> X:  DataFrame    </li>

    <li> random_state:  (seed)     </li>

````python
# Importing the class to perform train test split from model_selection module  
from sklearn.model_selection import train_test_split  
````

````python
  
# Separating the dependent and independent features  
X = irisDF.iloc[:, :-2].values  #        
y = irisDF.iloc[:, -1].values   #  
  
# Creating a testing dataset of size 0.2 times the whole dataset  
X_train, X_test, y_train, y_test = train_test_split(  X, y, 
                                                    test_size = 0.2, 
                                                    random_state = 10
                                                   )  

# Creating a testing dataset of size 0.2 times the whole dataset  
X_train, X_val, y_train, y_val = train_test_split(  X_train, y_train, 
                                                    test_size = 0.2, 
                                                    random_state = 10
                                                   )  

# Printing the shape of the training and testing dataset  
print("Shape of training set:")
print(X_train.shape)  
print(y_train.shape)  

print("Shape of validation set:")
print(X_val.shape)  
print(y_val.shape)  

print("Shape of test set:")
print(X_test.shape)  
print(y_test.shape)  
````

````python
from sklearn.metrics import accuracy_score  

# Finding accuracy by comparing actual response values(y_test) with predicted response value(y_pred)  
print("Validation set accuracy:", accuracy_score(y_val, y_pred_val))  

print("Test set accuracy:", accuracy_score(y_test, y_pred)) 
````

<h2 style="color:red; font-family:Nazanin;" dir="rtl">   (K)</h2>

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
    K       (cross validation)   K            K   K    .
</p> 

````python
acc = []
for i in range(1,11):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_val)
    acc.append(accuracy_score(y_val, pred_i))

acc
````

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
      K           K        .</p>

<p style="font-family:Nazanin; font-size:140%; text-align:right;" dir="rtl">
              K  .
</p> 

````python
plt.figure(figsize=(5, 4))

plt.plot(range(1, 11), acc, marker = "o")

plt.xticks(range(1,11))
plt.xlabel("K")
plt.ylabel("Validation set Accuracy")

plt.tight_layout()
plt.grid()
````

````python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score  

nn = MLPClassifier(hidden_layer_sizes = (10,15),
                   activation = 'relu',
                   solver = 'adam')

nn.fit(X_train, y_train)

y_pred = nn.predict(X_test)


print("\n Test set accuracy:", accuracy_score(y_test, y_pred)) 
````
